```python
def count_singer(songes):
    '''
    songes   所有歌曲的一个 pandas
    return   singers_count  
    '''
    singers_count={}   #歌手计数的字典
    for singer in songes.ix[:,[1]].itertuples(index=False):   #  itertuples 迭代成元祖    然后取下标为 0
        singer=singer[0]
        singers_count.setdefault(singer,0)
        singers_count[singer]+=1
    delname=[]

    for sing in singers_count.items():
        lines=sing[0].split('&')
        if len(lines) >=2:
            delname.append(sing[0])
    for sing in delname:
        lines=sing.split('&')
        if len(lines)>=2:
            for line in lines:
                singers_count.setdefault(line,0)
                singers_count[line]+=1
            del singers_count[sing]
    return singers_count
```

```python
singers_count=count_singer(songes)
singers_count
```

# 重名的歌曲

```python
def count_songes(songes):
    '''
    songes   所有歌曲的一个 pandas
    return   songes_count  
    '''
    songes_count={}  #重名歌曲集合
    for song in songes.ix[:].itertuples(index=False):   #  itertuples 迭代成元祖    然后取下标为 0
        name=song[2]
        songes_count.setdefault(name,{})
        songes_count[name].setdefault(song[1],0)
        songes_count[name][song[1]]+=1
        
    for song in songes.ix[:].itertuples(index=False):   #  itertuples 迭代成元祖    然后取下标为 0
        songname=song[2]
        singer=song[1]
        le=len(songes_count[songname])
        item=songes_count[songname][singer]
        if le==1 and item==1:
            del songes_count[songname]
    return songes_count
```

```python
songes_count=count_songes(songes)
songes_count
```

# 聚类模型

```python
#由于没有曲调 所以不能从调入手 那么对于英文的词分类就只能依靠于特定的词汇
```

```python
#在设置一个 stop_words 可以防止状态过拟合
import codecs
files=codecs.open('CNENstopwords.txt',encoding='utf8')
stop_words=[]
for line in files.readlines():
    stop_words.append(line.strip())
#stop_words
```

```python
import jieba
tags_count=[]
print(len(lyrics))
for lyric in lyrics:
    tags=jieba.lcut(lyric)
    judge_tag=[]
    for tag in tags:
        if tag not in stop_words:
            judge_tag.append(tag)
    strs=' '.join(judge_tag)
    tags_count.append(strs)
```

```python
tags_count[20]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer(analyzer="word")
transformer=TfidfTransformer()
tfidf=transformer.fit_transform(vectorizer.fit_transform(tags_count))
word=vectorizer.get_feature_names()
weight=tfidf.toarray()
tfidf
```

```python
len(word)  
weight[0][5585]
```

```python
from sklearn.cluster import MiniBatchKMeans
km = MiniBatchKMeans(n_clusters=40, init='k-means++', n_init=1,
                    init_size=200, batch_size=100)
'''
n_clusters    即我们的k值，一般需要多试一些值以获得较好的聚类效果
init          即初始值选择的方式，可以为完全随机选择'random',优化过的'k-means++'或者自己指定初始化的k个质心。一般建议使用默认的'k-means++'。
n_init        n_init则是每次用不一样的采样数据集来跑不同的初始化质心运行算法。
init_size     用来做质心初始值候选的样本个数，默认是batch_size的3倍，一般用默认值就可以了
batch_size    即用来跑Mini Batch KMeans算法的采样集的大小，默认是100.如果发现数据集的类别较多或者噪音点较多，需要增加这个值以达到较好的聚类效果。      
'''

s=km.fit(weight)
print(km.inertia_)   #评分  越低越好  距离
print(len(s.labels_))    #每个样本 被划分到的位置
print(s.cluster_centers_)  #查看质心
```

```python
num=0
cluster={}
for label in s.labels_:
    cluster.setdefault(label, ])
    cluster[label].append(num)  #这里 就可以知道 每首歌分属的集群了
    num+=1
```

```python
for item in cluster.items():
    list1=item[1]
    mark_one=0.0
    for i in list1:  #样本的索引
        for j in range(len(word)):
            mark_two=weight[i][j]
            if mark_two > mark_one:
                mark_one=mark_two
                index=j
    print(mark_one,'------',index)
    print('代表词为：',word[index])
    print('---------------')
```

```python
import pymysql
import pandas as pd
conn=pymysql.connect(host='localhost', port=3306,user='root',passwd='a',db='li',charset='utf8mb4')
try:
    with conn.cursor() as cursor:
        sql='select id,lrc from kuwo_music where id<1000 and id not in(select id from kuwo_music where lrc like \'%%%%暂无歌词%%%%\' and id<1000)'
        cursor.execute(sql)
        result=cursor.fetchall() 
finally:
    conn.close();
columns=['id','lrc']
songes=pd.DataFrame(list(result),columns=columns)

lyrics=songes['lrc']
#lyrics
#for lyric in lyrics:
    
```

```python
import re
numbers=0
for lyric in lyrics:
    result=re.findall(r'.*：.*?\,|.*:.*?\,',str(lyric))
    if(result!=[]):
        r1=str(lyric).replace(result[0],"").replace(']',"")
        lyrics[numbers]=r1
    else:
        result=re.findall(r'\[.*?',str(lyric))
        r1=str(lyric).replace(result[0],"").replace(']',"")
        lyrics[numbers]=r1
    numbers+=1
#songes_lyrics

word_count=[]
for lyric in lyrics:
    result=lyric.split(',')
    word_count+=result
    
word_count
```

```python
from sklearn.naive_bayes import MultinomialNB
import jieba.analyse
result={}
for lyric in word_count:
    tags=jieba.analyse.extract_tags(lyric,topK=5)
    result.setdefault(lyric,[])
    result[lyric]=tags
print(result)
```

```python
vocabSet = set([])
for value in result.values():
    vocabSet = vocabSet | set(value)     #  |  取并集 set的方法
vocabSet=list(vocabSet)
```

```python
vocabSet[0:20]
```

```python
import numpy as np
lyric_app=[]
for value in result.values():
    lyric_pack=np.zeros(len(vocabSet))   #创造一个  全部 0的词频矩阵 
    for val in value:
        if val in vocabSet:
            lyric_pack[vocabSet.index(val)]+=1
        lyric_app.append(lyric_pack)
target_ans=[]
for value in result.items():
    target_ans.append(value)
lyric_app=np.array(lyric_app)
```

```python
target_ans=np.array(target_ans)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB(alpha=1.0, fit_prior=False, class_prior=0)
mnb.fit(lyric_app,target_ans)
```
