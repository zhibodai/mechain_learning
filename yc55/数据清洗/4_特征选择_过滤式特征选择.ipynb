# 1.方差选择法：
使用方差作为特征评分标准，如果某个特征的取值差异不大，通常认为该特征对区分样本的贡献度不大
因此在构造特征过程中去掉方差小于阈值的特征


```python
#特征选择----方差选择法
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=1)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print('iris特征名称',iris.feature_names)
print('iris特征矩阵',iris.data)
```

```python
#特征选择--方差选择法
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=1) # threshold 为方差的阈值  默认0
selectedData=vt.fit_transform(iris.data)
print('方差选择法的特征：',selectedData)
```

```python
print(vt.variances_)
```

单变量特征选择：
采用皮尔逊相关系数（相关法）（回归问题）
使用 pearson 系数作为特征评分标准
相关系数绝对值越大，相关性越强(相关系数越接近于1或者-1时，相关性越强；越接近0越弱)
皮尔逊能够衡量线性相关的特征集

```python
#载入数据
from sklearn.datasets import load_iris
irisdata=load_iris()

#特征选择（皮尔逊）
from sklearn.feature_selection import SelectKBest
from scipy.stats import pearsonr  #计算皮尔逊相关系数
import numpy as np
```

```python
skb=SelectKBest(lambda X,Y: tuple(map(tuple,array(list(map(lambda x:pearsonr(x,Y),x.T) ) ).T ) ),k=2 )
newData=skb.fit_transform(irisdata.data,irisdata.target)
print(skb.scores_)
print(newData)
```
