```python
import numpy as np
import codecs
def loadDataSet(filename,delimitor='\t'):
    #打开文件
    file=codecs.open(filename,'r','utf-8')
    #针对文本中每一行的特征分割开来
    matrics=[line.strip().split(delimitor) for line in file.readlines()]
    #利用map函数 把每一行都转换为float函数
    data=[list(map(float,line)) for line in matrics]
    #将float的数据类型的值转化为矩阵返回
    return np.array(data)
```

```python
from sklearn.preprocessing import Imputer
data=loadDataSet('secom.data',' ')
'''
strategy:
    mean  特征列的 均值 替换
    median         中位数
    most_frequent 时   众数
'''
imp=Imputer(missing_values='NaN',strategy='mean',axis=0)   
imp.fit(data)  #模型
result=imp.transform(data)
print(result[0])
```
