```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x=np.linspace(-1,6,141)
plot_y=(plot_x-2.5)**2-1
```

```python
plt.plot(plot_x,plot_y)
plt.show()
```

```python
def dJ(theta):
    return 2*(theta-2.5)

def J(theta):
    return (theta-2.5)**2-1.
```

```python
theta=0.0
#下降步长
eta=0.1
#判断是否为最小值
epsilon=1e-8

while True:
    gradient=dJ(theta)
    lastTheta=theta
    theta=theta-eta*gradient
    
    if(abs(J(theta)-J(lastTheta))<epsilon):
        break

print(theta)
print(J(theta))
```

```python
theta=0.0
#学习率
eta=0.1
#判断是否为最小值
epsilon=1e-8
theta_history=[theta]


while True:
    gradient=dJ(theta)
    lastTheta=theta
    theta=theta-eta*gradient
    theta_history.append(theta)
    if(abs(J(theta)-J(lastTheta))<epsilon):
        break
        
plt.plot(plot_x,plot_y)
plt.plot(np.array(theta_history),J(np.array(theta_history) ),color='r',marker='+' )
plt.show()
```

```python
#封装上述函数
def gradient_descent(theta,eta,epsilon= 1e-8):
    theta_history=[theta]
    
    while True:
        gradient=dJ(theta)
        lastTheta=theta
        theta=theta-eta*gradient
        theta_history.append(theta)
        
        if(abs(J(theta)-J(lastTheta))<epsilon):
            break
    plt.plot(plot_x,plot_y)
    plt.plot(np.array(theta_history),J(np.array(theta_history) ),color='r',marker='+' )
    plt.show()
```

```python
eta=0.01
theta_history=[]
gradient_descent(theta=0.0,eta=eta)
```

```python
eta=0.8
theta_history=[]
gradient_descent(theta=0.0,eta=eta)
```

```python
eta=1.1
theta_history=[]
gradient_descent(theta=0.0,eta=eta)

#由于你的学习率太大直接跳过了极值点，进入了一个增大的过程最终导致python的compile崩溃
#所以进行算法优化
```

```python
def gradient_descent_2(theta,eta,n_iters=1e4,epsilon= 1e-8):
    '''
    n_iters  最大循环次数
    '''
    theta_history=[theta]
    i_iters=0
    
    while i_iters<n_iters:
        gradient=dJ(theta)
        lastTheta=theta
        theta=theta-eta*gradient
        theta_history.append(theta)
        
        
        if(abs(J(theta)-J(lastTheta))<epsilon):
            break
        i_iters+=1
    plt.plot(plot_x,plot_y)
    plt.plot(np.array(theta_history),J(np.array(theta_history) ),color='r',marker='+' )
    plt.show()
```

```python
eta=1.1
theta_history=[]
gradient_descent_2(theta=0.0,eta=eta,n_iters=10)
#数值太大了，学习率对于你的梯度算法而言重要性不言而喻    
#这里 five 的我任务我们可以theta = 0.01  对于大多数的函数基本适用
```
