```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
X=digits.data
y=digits.target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,stratify=digits.target)
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
%%time
knn_elf=KNeighborsClassifier()
knn_elf.fit(X_train,y_train)
```

```python
knn_elf.score(X_test,y_test)
```

```python
pca=PCA(n_components=2)

pca.fit(X_train)
X_train_reduction=pca.transform(X_train)
X_test_reduction=pca.transform(X_test)
```

```python
%%time
knn_elf=KNeighborsClassifier()
knn_elf.fit(X_train_reduction,y_train)
```

```python
knn_elf.score(X_test_reduction,y_test)
```

```python
pca.explained_variance_ratio_   # pca 对方差的解释程度 
```

```python
pca=PCA(n_components=X_train.shape[1])
pca.fit(X_train)
pca.explained_variance_ratio_    
```

```python
import numpy as np
#  x是 特征数量   y 表示pca 累和对于特征数量的值
plt.plot([i for i in range(X_train.shape[1])]
        ,[np.sum(pca.explained_variance_ratio_[:i+1]) for i in range(X_train.shape[1])])
plt.show()
```

```python
pca=PCA(0.95)
pca.fit(X_train)
```

```python
pca.n_components    # pca降维后的方差总和
pca.n_components_  # pca 降维后的特征数量
```

```python
X_train_reduction=pca.transform(X_train)
X_test_reduction=pca.transform(X_test)
```

```python
%%time

knn_elf=KNeighborsClassifier()
knn_elf.fit(X_train_reduction,y_train)
```

```python
knn_elf.score(X_test_reduction,y_test)  #这里PCA的效果就很明显了，准确率降低很少但时间确实几何倍的下降
```

```python
pca=PCA(n_components=2)
pca.fit(X)
X_reduction=pca.transform(X)
X_reduction.shape
```

```python
for i in range(10):  #手写数字 0-9  所以绘制9次 每次绘制的是一个数字
    plt.scatter(X_reduction[y==i,0],X_reduction[y==i,1],alpha=0.8)  # X_reduction[y==i,0] 其中0表示的你的维度的选择
plt.show()
```
