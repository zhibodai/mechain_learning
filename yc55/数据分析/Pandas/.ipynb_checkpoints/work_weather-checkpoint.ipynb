# 缺失值，异常值和重复值处理

```python
#1.查看是否的缺失值
df=pd.read_csv('E:/yc_55/hz_weather.csv')
df.head()
```

```python
df1=pd.pivot_table(df,values=['最高气温'],index=['天气'],columns=['风向'])
df1.isnull()
```

```python
#删除行
df1.dropna(axis=0)
```

```python
#删除列
df1.dropna(axis=1)
```

```python
#使用字符串代替
df1.fillna('missing')
```

```python
#使用前一个数据代替NaN  method=‘pad’
df1.fillna(method='pad')
```

```python
#使用后一个数据来代替NaN  method='bfill'   limit 可以限制代替的数目  下面限制每列只能代替一个
df1.fillna(method='bfill',limit=1)
```
