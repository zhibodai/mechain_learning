```python
#新增销售额索引   且等于 价格 乘  成交量

taobao['销售额']=taobao['价格']*taobao['成交量']
taobao[['销售额']]
```

```python
# 要求价格小于100  成交量大于10000

# 过滤
t1=taobao[ (taobao['价格']<100) & (taobao['成交量']>10000) ]
t1
```

```python
#将位置设为索引    按位置索引做排序
t1=taobao.set_index('位置')
t1=t1.sort_index()
t1.head()
```

```python
#按位置与卖家当索引   按位置来排序

t2=taobao.set_index(['位置','卖家']).sortlevel(0)   #sortlevel  根据第几个索引来排序来排序
t2.head()
```

```python
#  删除 ‘宝贝’，‘卖家’ ，按位置分组 求 ‘价格’  ，‘成交量’，‘销售额’的均值  按成交量排序

tao_mean=taobao.drop(['宝贝','卖家'],axis=1).groupby('位置').mean().sort_values('成交量',ascending=False)
tao_sum=taobao.drop(['宝贝','卖家'],axis=1).groupby('位置').sum().sort_values('成交量',ascending=False)
#axis=0  删除行   axis=1  删除列
# groupby  汇总
#sort_values 排序
tao_mean
```
