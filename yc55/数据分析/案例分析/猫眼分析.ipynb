```python
from pymongo import MongoClient
import json
from pyecharts import Geo
#使用 pyecharts 提供的map来生成一个按省/市来展示的地图
from pyecharts import Map
from pandas import pandas as pd

filmName='a_cool_fish'
client=MongoClient('localhost',27017)
db=client['li']
collection=db[filmName]

def showInChinesMap(list1):
    data=pd.DataFrame(list1)
    city=data.groupby(['city'])
    city_com=city['rate'].agg(['mean','count'])  #城市名  计数  
    city_com.reset_index(inplace=True)
    data_map=[(city_com['city'][i],city_com['count'][i]) for i in range(0,city_com.shape[0])]
    
    geo=Geo('GEO地理位置分析',title_pos='center',width=1200,height=800)
    while True:
        try:
            #cast 时有些城市是没有坐标的
            attr,val=geo.cast(data_map)
            #print(attr,val)
            geo.add("",attr,val,visual_range=[0,300],visual_text_color='#fff',symbol_size=10,is_visualmap=True,maptype='china')
        except ValueError as e:
            #使用pycharts提供的数据地图时  有一些猫眼数据城市找不到对应的坐标  所以在代码中 GEO添加出错的城市 直接删除 删选出来
            e=str(e).split('No coordinate is specified for ')[1]
            data_map=list(filter(lambda item:item[0] !=e,data_map))  #[(城市名,评分)]
        else:
            break
    geo.render('geo_city_localtion.html')
    
if __name__=='__main__':
    showInChinesMap(list(collection.find()))
```

```python
from pymongo import MongoClient
import json
from pyecharts import Geo
#使用 pyecharts 提供的map来生成一个按省/市来展示的地图
from pyecharts import Map
from pandas import pandas as pd

filmName='a_cool_fish'
client=MongoClient('localhost',27017)
db=client['li']
collection=db[filmName]

#将每个市的名字换成它所对应的省份名
def getRealName(name,jsonObj):
    for province in jsonObj['provinces']:
        for city in province['citys']:  #citys集合
            if city['citysName'].startswith(name):
                return province['provinceName']  #如果这个 name 所代表的市在json中有  则返回它所在的省名
    return name       #没有则返回本名


def realKeys(name):
    return name.replace(u"省","").replace(u"市","").replace(u"回族自治区","").replace(u'维吾尔自治区',"").replace(u"壮族自治区","").replace(u"自治区","")

def getCountyData(list1):
    data=pd.DataFrame(list1)
    city=data.groupby(['city'])
    
    city_com=city['rate'].agg(['mean','count'])
    
    city_com.reset_index(inplace=True)
    fo=open('citys.json','r',encoding='utf8')
    citys_info=fo.readlines()   #list列表[0]
    
    citysJson=json.loads(str(citys_info[0] ) ) 
    
    data_map_all=[(getRealName(city_com['city'][i],citysJson),city_com['count'][i])
                 for i in range(0,city_com.shape[0])]
    
    data_map_list={}
    for item in data_map_all:
        if item[0] in data_map_list:
            
            value = data_map_list[item[0]]
            value +=item[1]
            data_map_list[item[0]]=value
        else:
            data_map_list[item[0]]=item[1]
            
    data_map=[(realKeys(key),data_map_list[key]) for key in data_map_list.keys()]
    
    return data_map

def generateMap(data_map):
    map=Map('省份评论数',width=1200,height=800,title_pos='center')
    geo = Geo('geo地理位置分析',title_pos='center',width=1200,height=800)
    while True:
        try:
            attr,val=geo.cast(data_map)
            map.add("",attr,val,visual_range=[0,800],visual_text_color='#fff',symbol_size=5,
                   is_visualmap=True,maptype='china',is_map_symbol_show=False,is_label_show=True,is_roam=False)
        except ValueError as e:
            e=str(e).split('No coordinate is specified for ')[1]
            data_map=filter(lambda item:item[0] !=e,data_map)
        else:
            break
    map.render("city_rate_count.html")
    
if __name__=='__main__':
    data_map=getCountyData(list(collection.find()))
    generateMap(data_map)
```

```python
from pymongo import MongoClient
import json
from pyecharts import Geo
#使用 Pyecharts提供的map来生成一个按省、市来展示的地图
from pyecharts import Map
from pyecharts import Bar
from pandas import pandas as pd

'''
可视化一下每一个评分的人数，这里用柱状图来展示
'''
fileName='a_cool_fish'
client=MongoClient('localhost',27017)
db=client['li']
collection=db[fileName]

def showPersonNumsByScore(list1):
    data=pd.DataFrame(list1)
    #按评分分类
    rateData=data.groupby(['rate'])
    rateDataCount=rateData['rate'].agg(["count"])
    rateDataCount.reset_index(inplace=True)
    print(rateDataCount)
    #按评分从高到低来排列
    count=rateDataCount.shape[0]-1
    attr=[rateDataCount["rate"][count-1] for i in range(0,rateDataCount.shape[0]) ]
    v1 = [rateDataCount["count"][count-1] for i in range(0,rateDataCount.shape[0]) ]
    bar=Bar('各等级数量')
    bar.add('数量',attr,v1)
    bar.render("rate_date.html")
    
if __name__=="__main__":
    showPersonNumsByScore(list(collection.find()))
    
```

```python
from pymongo import MongoClient
'''
pip install wordcloud
'''
import jieba
import matplotlib as plt
import numpy as np
import image
from  wordcloud import WordCloud,STOPWORDS
from scipy.misc import imread
import pandas as pd

fileName='a_cool_fish'
client=MongoClient('localhost',27017)
db=client['li']
collection=db[fileName]

def showWordCould(list1):
    data = pd.DataFrame(list1)
    
    comment=jieba.cut(str(data['comment']),cut_all=False)
    
    wl_space_split=" ".join(comment)
    
    backgroupImage=imread('bg.png')
    
    stopword=STOPWORDS.copy()
    
    wc=WordCloud(width=1920,height=1080,background_color='white',
                 mask=backgroupImage,
                 font_path="'C:/Windows/Fonts/simkai.ttf",
                 stopwords=stopword,max_font_size=400,
                 random_state=50)
    wc.generate_from_text(wl_space_split)
    wc.to_file('unknow_word_cloud.png')
    
if __name__=="__main__":
    showWordCould(list(collection.find()))
```
