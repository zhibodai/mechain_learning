```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#解决中文问题
plt.rcParams['font.sans-serif']=['simHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#加载数据
data=pd.read_csv('Mountains.csv')
data.head()
```

```python
#查看数据  缺失值  数据的条数  列数 类型
data.info()
```

```python
data.describe()
```

#  数据清洗

```python
#重命名列表  以便后序操作 data-》 dataframe
data.rename(columns={'Height (m)':'Height','Ascents bef. 2004':'Success','Failed attempts bef. 2004':'Failed'},inplace=True)
#  inplace  在原数据上进行替换
data.head()
```

```python
#清洗
#1.Failed列中的空值填为0
data['Failed']=data['Failed'].fillna(0).astype(int)

#定义一个函数处理success列中的特殊值
def process_success(val):
    if '>' in str(val):
        return 200
    elif 'Many' in str(val):
        return 150
    else:
        return val
    
data['Success']=data['Success'].apply(process_success)
data['Success']=data['Success'].fillna(0).astype(int)

# First ascent 列中的unclimbed 数据行 删除
data=data[ data['First ascent']!='unclimbed']
data['First ascent']=data['First ascent'].astype(int)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#1.每年登顶的次数
plt.hist(data['First ascent'].astype(int),bins=20)
plt.ylabel('首次登顶的高峰数量')
plt.xlabel('年份')
plt.title('首次登顶次数与年份之间的关系')
plt.show()
```

```python
#高峰与海拔
plt.hist(data['Height'].astype('int'),bins=20)
plt.ylabel('高峰数量')
plt.xlabel('海拔高度')
plt.title('高峰_海拔')
plt.show()
```

```python
#海拔与登顶的数量
data['Attempts']=data['Failed']+data['Success']
fig=plt.figure(figsize=(13,7))
fig.add_subplot(211)
plt.scatter(data['First ascent'],data['Height'],c=data['Attempts'],alpha=0.8,s=50)
plt.xlabel('登顶')
plt.ylabel('海拔')

fig.add_subplot(212)
plt.scatter(data['First ascent'],data["Rank"].max()-data['Rank'],c=data['Attempts'],alpha=0.8,s=50)
plt.ylabel('排名')
plt.xlabel('登顶')
plt.show()
```
