员工离职原因分析
项目任务：分析离职原因与各数据之间的关系

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#读取数据 显示前几行
sample_data=pd.read_csv('train.csv')
sample_data.head()
```

```python
#数据集的情况 ：数据条数(空值数，非空数)
sample_data.info()
```

```python
#统计数据
sample_data.describe()
```

```python
#正负样本比列（离职：在职）
pos_data=sample_data[sample_data["Attrition"]==1 ]
neg_data=sample_data[sample_data["Attrition"]==0 ]
pos_len=len(pos_data)
neg_len=len(neg_data)

print("正样品记录数:{},所占的比列:{}".format(pos_len,pos_len/(pos_len+neg_len)))
print("负样品记录数:{},所占的比列:{}".format(neg_len,neg_len/(pos_len+neg_len)))
```

```python
#用图表来显示  离职与教育程度的关系
plt.figure()
sns.countplot(x='Attrition',data=sample_data,hue='Education')
```

```python
#以多图表显示单个变量的情况
plt.figure(figsize=(18,10) )
'''
seaborn.boxplot(x=None, y=None, hue=None, data=None, 
order=None, hue_order=None, orient=None, color=None, palette=None, 
saturation=0.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5, notch=False, ax=None, **kwargs)

图形解读
    异常点
    最大值
    
    上四分位数
    中位数
    下四分位数
    
    最小值
    异常值
'''
#教育的影响
plt.subplot(2,3,1)
sns.boxplot(x='Attrition',y='Education',data=sample_data)
#上班与距离的影响
plt.subplot(2,3,2)
sns.boxplot(x='Attrition',y='DistanceFromHome',data=sample_data)
#上班与月收入的影响
plt.subplot(2,3,3)
sns.boxplot(x='Attrition',y='MonthlyIncome',data=sample_data)
#上班与服务过的公司的影响
plt.subplot(2,3,4)
sns.boxplot(x='Attrition',y='NumCompaniesWorked',data=sample_data)

plt.subplot(2,3,5)
sns.boxplot(x='Attrition',y='EmployeeNumber',data=sample_data)
```

```python
plt.figure()
#婚姻状况的影响
plt.subplot(2,1,1)
sns.countplot(x="Attrition",hue='MaritalStatus',data=sample_data)
#性别的影响
plt.subplot(2,1,2)
sns.countplot(x='Attrition',hue='Gender',data=sample_data)
```
