```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import scipy as sp
from scipy import stats
%matplotlib inline
plt.rcParams['font.sans-serif']=['simHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
data=pd.read_csv('vgsales.csv')
data.head()
```

```python
#按类别统计 求和
GenreSum=data.groupby('Genre').sum().reset_index()
GenreSum
```

```python
plt.figure(figsize=(10,5) )
ax=sns.barplot(x=GenreSum['Genre'],y=GenreSum['Global_Sales'].astype(int))
ax.set_xlabel(xlabel='游戏类型')
ax.set_ylabel(ylabel='全球总销售量')
ax.set_title(label='畅销的游戏类型')
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(x=GenreSum['Genre'],y=GenreSum['NA_Sales'].astype(int))
ax.set_xlabel(xlabel='游戏类型')
ax.set_ylabel(ylabel='北美销售量')
ax.set_title(label='北美销售的游戏类型')

plt.figure(figsize=(10,5))
ax=sns.barplot(x=GenreSum['Genre'],y=GenreSum['EU_Sales'].astype(int))
ax.set_xlabel(xlabel='游戏类型')
ax.set_ylabel(ylabel='欧洲销售量')
ax.set_title(label='欧洲销售的游戏类型')

plt.figure(figsize=(10,5))
ax=sns.barplot(x=GenreSum['Genre'],y=GenreSum['JP_Sales'].astype(int))
ax.set_xlabel(xlabel='游戏类型')
ax.set_ylabel(ylabel='日本销售量')
ax.set_title(label='日本销售的游戏类型')

plt.figure(figsize=(10,5))
ax=sns.barplot(x=GenreSum['Genre'],y=GenreSum['Other_Sales'].astype(int))
ax.set_xlabel(xlabel='游戏类型')
ax.set_ylabel(ylabel='其他销售量')
ax.set_title(label='其他销售的游戏类型')

plt.show()
```

```python
GenreSumNA=round(GenreSum['NA_Sales'].sum(),2)
GenreSumEU=round(GenreSum['EU_Sales'].sum(),2)
GenreSumJP=round(GenreSum['JP_Sales'].sum(),2)
GenreSumOther=round(GenreSum['Other_Sales'].sum(),2)
GenreSumCount=GenreSum['Global_Sales'].sum()
sales=[GenreSumNA,GenreSumEU,GenreSumJP,GenreSumOther]
Name=['NA_Sales','EU_Sales','JP_Sales','Other_Sales']
plt.title('全球各地区销售量')
colors=['yellow','red','green','blue']
explode=[0.2,0,0,0]

#pie(数据 标签 explode偏移量)
plt.pie(sales,labels=Name,colors=colors,explode=explode)
```

```python
datayear=data.groupby('Year').sum().reset_index()   #如果不用reset_index() 的话，则‘Year’这个列就会被生成行索引
datayear=datayear[datayear['Year']!=2020.0]
datayear.head()
```

```python
plt.plot(datayear['Year'].astype(int),datayear['Global_Sales'])
plt.xlabel('年份')
plt.ylabel('总销售量')
plt.title('每年的销售量')
plt.show()
```

```python
plt.plot(datayear['Year'].astype(int),datayear['NA_Sales'],'b',datayear['Year'].astype(int),datayear['EU_Sales'],'y',datayear['Year'].astype(int),datayear['JP_Sales'],'r',datayear['Year'].astype(int),datayear['Other_Sales'],'g')
plt.legend(['NA_Sales','EU_Sales','JP_Sales','Other_Sales'],loc=2)  #legend绘制图例    loc= 1 - 8  方位
plt.xlabel('年份')
plt.ylabel('销售量')
plt.title('每年各地的销售量')
plt.show()
```

```python
#按照发行商的全球销售量排序  组成一个DataFrame
dataPublish=data.groupby('Publisher').sum().reset_index()
dataPublish=pd.DataFrame(dataPublish.sort_values(by='Global_Sales',ascending=False) )[0:10]
dataPublish
```

```python
def GenreName(name):
    dataGenre=data[data["Genre"]==name]
    dataGenreName=dataGenre['Name']   #取游戏名
    dataGenreName.to_csv(name+'.txt',index=False)
```

```python
#词云  
#导入 wordcloud模块
from wordcloud import WordCloud
from scipy.misc import imread
def wordCloudName(name,image):
    #读取一个txt文件
    text=open(name,'r',encoding='gbk').read()
    #读入背景图片
    bg_pic=imread(image+'.png')
    #生成词云
    wordcloud=WordCloud(width=400,height=900,mask=bg_pic,background_color='white',max_words=1000,scale=3.0).generate(text)
    #显示
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
wordCloudName('Sports.txt','bg')
```
