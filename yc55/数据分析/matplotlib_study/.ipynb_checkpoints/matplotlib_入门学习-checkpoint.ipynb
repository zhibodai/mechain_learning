```python
# pyplot使用
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])  #[1,2,3,4] 对应y值 x值为0,1,2,3
plt.show()
```

```python
#设置图形的属性
#1.指定x值 及颜色
plt.plot([1,2,3,4],[1,4,9,16],'bo')
plt.show()
```

```python
# 2.刻度与标题
plt.axis( [0,5,0,20] )  #axis() 指定刻度  [x最小，x最大，y最小，y最大]
plt.title('hello world')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()
```

```python
#3. 结合numpy 完成绘图：在一张图中绘制三种不同的趋势
import math
import numpy as np
x=np.arange(0,2.5,0.1)
y1=list( map(math.sin,math.pi*x) )
y2=list( map(math.sin,math.pi*x+math.pi/2))
y3=list( map(math.sin,math.pi*x-math.pi/2))

# plt.plot(x,y1,'b*',x,y2,'g',x,y3,'ys')
plt.plot(x,y1,'b*')
plt.plot(x,y2,'g^')
plt.plot(x,y3,'ys')
plt.show()
```

```python
#4.也可以用点和线来组成不同的线型
plt.plot(x,y1,'b----',x,y2,'g',x,y3,'r-.-.')
plt.show()
```

```python
#5.使用kwargs 指定关键字参数
#语法 plot(*args,**kwargs)
plt.plot(x,y1,'b--',x,y2,'g',x,y3,'r-.',linewidth=3)
plt.show()
```

```python
#绘制多个图形
#subplot（*，*，*） 三个参数都是数字 第一个决定垂直方向分为几个部分，第二个决定水平方向几个部分  第三个决定是第几张图
x=np.arange(0,2.5,0.1)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)
plt.subplot(121)      #subplot(行数列数第几个图形)
plt.plot(x,y1,'b-')
plt.subplot(122)
plt.plot(x,y2,'r--')
plt.show()
```

```python
#分左右 
plt.figure(figsize=(40,10) )
plt.subplot(121)
plt.plot(x,y1,'b-')
plt.subplot(122)
plt.plot(x,y2,'r--')
plt.show()
```

```python
#添加文本  -> 坐标轴标签xlabel() ylabel()  title()  指定属性
#在图表中任何位置添加文本  text（x,y,fontdict=None,**kwargs）
plt.axis([0,5,0,20])  #指定刻度的范围
#title('标题',fontsize='标题的大小',fontname='字体名')
plt.title('hello world',fontsize=20,fontname='Times New Roman')
# 关于字体，如果使用fontname加载系统默认字体，可以直接输入字体名
#但如果是第三方字体，则要创建字体管理
plt.xlabel('counting')
plt.ylabel('squre values')
plt.text(1,1.5,'First')    #图形 文本
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()
```

```python
#中文文本的解决方案：
#matplot 显示图例问题的中文问题         https://segmentfault.com/a/1190000005144275
import matplotlib.font_manager as fm
myfont=fm.FontProperties(fname='C:/Windows/Fonts/simkai.ttf') 
plt.axis([0,5,0,20])
plt.title('我的第一个图表',fontsize=20,fontname='Times New Roman',fontproperties=myfont)
plt.xlabel('计数',fontproperties=myfont)
plt.ylabel('平方值',fontproperties=myfont)
plt.text(1,1.5,'一',fontproperties=myfont)
plt.text(2,4.5,'二',fontproperties=myfont)
plt.text(3,9.5,'三',fontproperties=myfont)
plt.text(4,16.5,'四',fontproperties=myfont)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()
```

```python
#加入公式  plt.text(1.1,12,r'$***$,fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})    ****是数学公式   alpha透明度
#matplotlib支持数学表达式整合了laTex表达式
#将表达式内容置于2个$符号之间  就可以在文本里面进行添加laTex  通常在包含laTex字符前添加r字符
import matplotlib.font_manager as fm
myfont=fm.FontProperties(fname='C:/Windows/Fonts/simkai.ttf')
plt.axis([0,5,0,20])
plt.title('我的第一个图表',fontsize=20,fontname='Times New Roman',fontproperties=myfont)
plt.xlabel('计数',fontproperties=myfont)
plt.ylabel('平方值',fontproperties=myfont)
plt.text(1,1.5,'一',fontproperties=myfont)
plt.text(2,4.5,'二',fontproperties=myfont)
plt.text(3,9.5,'三',fontproperties=myfont)
plt.text(4,16.5,'四',fontproperties=myfont)

#指定图例  显示文本（x,y,内容）
# $***$   ***数学公式
plt.text(1.1,12,r'$y=x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.5})

plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()
```

```python
#加入网格  grid()
#加入图例说明  plt.legend(['图例'])
import matplotlib.font_manager as fm
myfont=fm.FontProperties(fname='C:/Windows/Fonts/simkai.ttf')
plt.axis([0,5,0,20])
plt.title('我的第一张表',fontsize=20,fontname='Times New Roman',fontproperties=myfont)
plt.xlabel('计数',fontproperties=myfont)
plt.ylabel('平方值',fontproperties=myfont)
plt.text(1,1.5,'一',fontproperties=myfont)
plt.text(2,4.5,'二',fontproperties=myfont)
plt.text(3,9.5,'三',fontproperties=myfont)
plt.text(4,16.5,'四',fontproperties=myfont)
plt.text(1.1,12,r'$y=x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.5})
plt.grid(True)  #设置网格  默认False

plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[16,9,4,1],'g*')

plt.legend(['series','second'],loc=0)   #loc表示位置  由0-10的值组成
plt.show()
```
