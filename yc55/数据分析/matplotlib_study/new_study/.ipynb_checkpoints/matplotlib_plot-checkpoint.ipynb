```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-1,1,50)  # 在指定的间隔（-1,1）内返回均匀间隔的数字。（50个）
y=2*x+1
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-1,1,50)  # 在指定的间隔（-1,1）内返回均匀间隔的数字。（50个）
y=x**2+1
plt.plot(x,y)
plt.show()
```

# 多张图展示

```python
x=np.linspace(-3,3,50)
y1=x*2+1
y2=x**2
plt.figure(figsize=(8,5)) # figsize 展示框的大小
plt.plot(x,y1)
plt.plot(x,y2,linestyle='--')  # linestyle 线的形状

plt.figure()
plt.xlim(-1,2)   # x轴的坐标限制
plt.ylim(0,3)    # y轴的坐标限制

plt.xticks(np.linspace(-0.5,1.5,5))  #要求x展示那几个坐标
#plt.yticks([0,1,2],['bad','just','pretty'])  # 将y轴坐标的对应的展示值替代
plt.yticks([0,1,2],[r'$bad$',r'$just\ so$',r'$pretty\ \alpha$'])  # $ 将字体替换成好看的字体，某些专业的符号 可以在前面加上 \  进行机器识别

plt.xlabel('I am X')  # 对于X轴的描述
plt.ylabel('I am Y')  # 对于Y轴的描述
plt.plot(x,y2)    # plot 折现图
plt.show()
```

```python
plt.figure()
plt.xlim(-1,2)   # x轴的坐标限制
plt.ylim(0,3)    # y轴的坐标限制

plt.xticks(np.linspace(-0.5,1.5,5))  #要求x展示那几个坐标
#plt.yticks([0,1,2],['bad','just','pretty'])  # 将y轴坐标的对应的展示值替代
plt.yticks([0,1,2],[r'$bad$',r'$just\ so$',r'$pretty\ \alpha$'])  # $ 将字体替换成好看的字体，某些专业的符号 可以在前面加上 \  进行机器识别

plt.xlabel('I am X')  # 对于X轴的描述
plt.ylabel('I am Y')  # 对于Y轴的描述
plt.plot(x,y2)    # plot 折现图



# gca = 'get current axis'
ax=plt.gca()
'''
首先将 不需要的边框颜色去掉  ，然后设置边框成为需要的x，y轴，最后确定，x，y轴的位置
'''
ax.spines['right'].set_color('none')  # 设置 指定 边框消失
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position( ('data',0) )  # 设置 x 轴的中心在 0  
ax.spines['left'].set_position( ('data',0) )  # 设置 y 轴的中心在 0

plt.show()
```

```python
plt.figure()
plt.xlim(-1,2)   # x轴的坐标限制
plt.ylim(0,3)    # y轴的坐标限制
plt.xlabel('I am X')  # 对于X轴的描述
plt.ylabel('I am Y')  # 对于Y轴的描述


plt.xticks(np.linspace(-0.5,1.5,5))  #要求x展示那几个坐标
#plt.yticks([0,1,2],['bad','just','pretty'])  # 将y轴坐标的对应的展示值替代
plt.yticks([0,1,2],[r'$bad$',r'$just\ so$',r'$pretty\ \alpha$'])  # $ 将字体替换成好看的字体，某些专业的符号 可以在前面加上 \  进行机器识别


plt.plot(x,y1,label='up')    # plot 折现图
plt.plot(x,y2,linestyle='--',label='down')
plt.legend()  #显示图例

plt.show()
```

# 添加注解

```python
x=np.linspace(-3,3,50)  # 在指定的间隔（-1,1）内返回均匀间隔的数字。（50个）
y=2*x+1

plt.plot(x,y)
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position( ('data',0) )  # 设置 x 轴的中心在 0  
ax.spines['left'].set_position( ('data',0) )  # 设置 y 轴的中心在 0

#注解
x0=1
y0=2*x0+1
plt.scatter(x0,y0,color='r')
plt.plot([x0,x0],[y0,0],'k--')  #  这里前面的 [ x的点]  [y的点]  进行对应

plt.annotate(r'$2x+1=3$',xy=(x0,y0))  # xy=(x0,y0)打印到那个位置
plt.text(-3.7,3,r'$This\ is\ the\ some\ text\ mu\ \sigma_i$')


plt.savefig('chart.jpg')  #转为图片下载
```
