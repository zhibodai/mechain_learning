# scatter

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X)  # color value

plt.scatter(X,Y,s=75,c=T,alpha=0.5)
plt.show()
```

```python
n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X)  # color value

plt.scatter(X,Y,s=75,c=T,alpha=0.5)
plt.xlim( (-1.5,1.5) )
plt.ylim( (-1.5,1.5) )
#隐藏 x y 的坐标轴
plt.xticks( () )
plt.yticks( () )

plt.show()
```

# bar

```python
n=12
X=np.arange(n)
print(X)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,color='#9999ff',edgecolor='white')
plt.bar(X,-Y1,color='#ff9999',edgecolor='white')

#给每个 柱状图的 text  在上面绘制出 y的值
for x,y in zip(X,Y1):
    plt.text(x+0.4,y+0.05,'%.2f'%y,ha='center',va='bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4,-y-0.05,'%.2f'%y,ha='center',va='top')  

    
plt.xlim(-1.5,n)
plt.xticks( () )
plt.ylim(-1.25,1.25)
plt.yticks( () )

plt.show()
```
