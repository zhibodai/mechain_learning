```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
import scipy as sp
```

```python
#创建figure对象pl
fig=plt.figure()
'''
plt.subplot(231)
plt.hist
'''
#fig 是一个图像实例，采用这个实例的方法来绘图  前面采用plt静态方法绘图
ax1 = fig.add_subplot(2,3,1)  # 2行 3列 第一个
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)

#第一个例子只有plot（）
  #绘制直方图  bins：几块
ax1.hist( np.random.randn(100),bins=10,color='b',alpha=0.3)

#绘制带线的直方图
x=np.linspace(-5,15,50)   #从-5  到  15 随机产生 50个数
#pdf :随机产生的概率密度函数  产生对应的x的这种分布的y   绘制高斯分布
ax2.plot(x,sp.stats.norm.pdf(x=x,loc=5,scale=2) )
#产生服从这种分布的样本  对随机变量进行随机取值，可以通过size参数指定输出的数组大小
ax2.hist(sp.stats.norm.rvs(loc=5,scale=2,size=200),bins=50,normed=True,color='red',alpha=0.5)

# 绘制散点图
x=np.arange(50)
y=x+5*np.random.rand(50)
ax3.scatter(x,y)

#柱状图
x=np.arange(5)
y1,y2=np.random.randint(1,25,size=(2,5) )
width=0.25
ax4.bar(x,y1,width,color='r')
ax4.bar(x+width,y2,width,color='g')
ax4.set_xticks(x+width)   # x轴的刻度的位置
ax4.set_xticklabels(['a','b','c','d','e'])  #x轴的刻度
plt.show()
```

```python
#条状图
index=np.arange(5)
values=[5,7,3,4,6]
plt.title('bar chart')
plt.bar(index,values)
plt.xticks(index+0.4,['a','b','c','d','e'])
plt.show()
```

```python
#水平条状图
index=np.arange(5)
values=[5,7,3,4,6]
plt.title('bar chart')
plt.barh(index,values)
plt.yticks(index+0.4,['a','b','c','d','e'])
plt.show()
```

```python
#多序列条状图
index=np.arange(5)+0.3
values1=[5,7,3,4,6]
values2=[4,6,4,5,7]
values3=[3,6,5,4,6]

bw=0.3  #bar的宽度
plt.axis([0,5,0,8])
plt.title('multi series bar char',fontsize=20)
plt.bar(index,values1,bw,color='b')
plt.bar(index,values2,bw,color='g')
plt.bar(index,values3,bw,color='r')
plt.xticks(index,['a','b','c','d','e'])
plt.show()
```

```python
#水平多序列条状图
index=np.arange(5)+0.3
values1=[5,7,3,4,6]
values2=[4,6,4,5,7]
values3=[3,6,5,4,6]

bw=0.3  #bar的宽度
plt.axis([0,8,0,5])
plt.title('multi series bar char',fontsize=20)
plt.barh(index,values1,bw,color='b')
plt.barh(index,values2,bw,color='g')
plt.barh(index,values3,bw,color='r')
plt.yticks(index,['a','b','c','d','e'])
plt.show()
```

```python
#多序列堆积条状图   width   bottom
series1=np.array([3,4,5,3])
series2=np.array([1,2,2,5])
series3=np.array([2,3,3,4])

bw=0.5
index=np.arange(4)+0.5
plt.axis([0,4,0,14])
plt.bar(index,series1,color='r',width=bw)
plt.bar(index,series2,color='b',width=bw,bottom=series1)
plt.bar(index,series3,color='g',width=bw,bottom=(series1+series2))
plt.xticks(index,['one','two','three','four'])
plt.show()
```

```python
#水平多序列堆积条状图      left   height
series1=np.array([3,4,5,3])
series2=np.array([1,2,2,5])
series3=np.array([2,3,3,4])

bw=0.5
index=np.arange(4)+0.5
plt.axis([0,14,0,4])
plt.barh(index,series1,color='r',height=bw)
plt.barh(index,series2,color='b',height=bw,left=series1)
plt.barh(index,series3,color='g',height=bw,left=(series1+series2))
plt.yticks(index,['one','two','three','four'])
plt.show()
```

```python
#还可以用不同的影线来填充     1.将条状图颜色设置颜色  hatch关键字参数来指定影线类型  使用 |  /  \  - * 来表示
#同一个符号越多越密集  
series1=np.array([3,4,5,3])
series2=np.array([1,2,2,5])
series3=np.array([2,3,3,4])

bw=0.5
index=np.arange(4)+0.5
plt.axis([0,4,0,14])
plt.bar(index,series1,color='w',width=bw,hatch='xx')
plt.bar(index,series2,color='w',width=bw,bottom=series1,hatch='/////')
plt.bar(index,series3,color='w',width=bw,bottom=(series1+series2),hatch='\\\\\\')
plt.xticks(index,['one','two','three','four'])
plt.show()
```

```python
#对比关系图
x0=np.arange(9)
y1=np.array([1,3,4,6,6,4,3,2,7])
y2=np.array([2,1,4,5,8,7,1,3,4])
plt.ylim(-8,8)
plt.bar(x0,y1,0.5,color='r',edgecolor='w')
plt.bar(x0,-y2,0.5,color='b',edgecolor='w')
plt.xticks(())
plt.grid(True)
for x,y in zip(x0,y1):
    plt.text(x,y+.05,'%d' %y,ha='center',va='bottom')
for x,y in zip(x0,y2):
    plt.text(x,-y-.05,'%d' %y,ha='center',va='top')
    
plt.show()
```

```python
#饼图  pie()
labels=['Nokia','Samsung','Apple','Lumia']
value=[10,30,45,15]
colors=['yellow','green','red','blue']
explode=[0.2,0,0,0]  #抽取圆饼中的一块
plt.pie(value,labels=labels,colors=colors,explode=explode,startangle=180)  #explode对每一个快进行位置设定
plt.axis('equal')  #正圆
plt.title('A Pie Chart')
plt.show()
```
