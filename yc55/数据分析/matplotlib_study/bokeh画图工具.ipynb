```python
from bokeh.plotting import figure,output_notebook,show
output_notebook()
p=figure(title='折线图',plot_width=300,plot_height=300)
p.line(x=[1,2,3,4,5],y=[6,7,2,4,5])
show(p)
```

```python
from bokeh.plotting import figure,output_notebook,show
output_notebook()
p=figure(plot_width=300,plot_height=300)
p.multi_line(xs=[[1,2,3],[2,3,4]],ys=[[6,7,2],[4,5,7]],color=['red','green'])

show(p)
```

```python
from bokeh.plotting import figure,output_notebook,show
output_notebook()
plot=figure(plot_width=300,plot_height=300,title='柱状图')
plot.vbar(x=[1,2,3],width=0.5,bottom=0,top=[1,2,3],color='#CAB2D6')

show(plot)
```
